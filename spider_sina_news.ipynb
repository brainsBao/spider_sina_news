```python
import requests
from bs4 import BeautifulSoup
import copy
import re
import json

contentNew='http://news.sina.com.cn/c/nd/2018-03-06/doc-ifxipenn4806584.shtml'
u='http://news.sina.com.cn/china/'
cUrl='http://comment5.news.sina.com.cn/page/info?version=1&format=js&channel=gn&newsid=comos-{}&group=0&compress=0&ie=gbk&oe=gbk&page=1&page_size=20&jsvar=loader_1520306824135_42700275'

#根据分页js中的非同步载入的api。。。连接，获取每个新闻y页面的网页
def getSplitPageUrl(api_url):
    splitUrl=[]
    res=requests.get(api_url)
    jd=json.loads(res.text.lstrip('  newsloadercallback(').rstrip(');'))
    for link in jd['result']['data']:
        splitUrl.append(link['url'])
    return splitUrl

#输入一个需要爬虫的网址
def inputUrl():
    url=input('website: ')
    return url
#对输入网址进行基本解析 为soup
def getReady(url):
    res=requests.get(url)
    res.encoding='utf-8'
    soup=BeautifulSoup(res.text,'html.parser')
    return soup
#获取所有文章内容 包括时间地点正文作者等
def getAllArticleInfo():
    result={}
    all_result=[]
    news=getArticleUrl()
    for new in news:
        soup=getReady(new)
        title=soup.select('.main-title')[0].text
        article=soup.select('#article')[0].text.strip()
        datetime=soup.select('.date')[0].text.strip('\n\t')
        result['title']=title
        result['article']=article
        result['datetime']=datetime
        all_result.append(result)
        all_result=copy.deepcopy(all_result)
    return all_result
#获取评论数获取
def commentObtain():
    links=getArticleUrl()
    totalCommentsCount=[]
    for link in links:
        id=re.search('doc-i(.+).shtml',link.split('/')[-1]).group(1)
        commentUrl=cUrl.format(id)
        res=requests.get(commentUrl).text
        begin_cut=re.search('(.+?)=',res).group(0)
        res=res.lstrip(begin_cut)
        jd=json.loads(res)
        totalCommentsCount.append(jd['result']['count']['show'])
    return totalCommentsCount
#获取页面上新闻的url   
def getArticleUrl():
    url=inputUrl()
    a_label_list=[]
    soup=getReady(url)
    for a_label in soup.select('.blk122 a'):
        a_label_list.append(a_label['href'])
    return a_label_list

```

```python
import requests
import json
import pandas
import sqlite3
def getNewsContent(url):
    all_result=[]
    result={}
    for url in getSplitPageUrl(url):
        res=requests.get(url)
        res.encoding='utf-8'
        soup=BeautifulSoup(res.text,'html.parser')
        title=soup.select('.main-title')[0].text
        article=soup.select('#article')[0].text.strip()
        datetime=soup.select('.date')[0].text.strip('\n\t')
        result['title']=title
        result['article']=article
        result['datetime']=datetime
        all_result.append(result)
        all_result=copy.deepcopy(all_result)
    return all_result
    
    
def getSplitPageUrl(api_url):
    splitUrl=[]
    res=requests.get(api_url)
    jd=json.loads(res.text.lstrip('  newsloadercallback(').rstrip(');'))
    for link in jd['result']['data']:
        splitUrl.append(link['url'])
    return splitUrl

url='http://api.roll.news.sina.com.cn/zt_list?channel=news&cat_1=gnxw&cat_2==gdxw1||=gatxw||=zs-pl||=mtjj&level==1||=2&show_ext=1&show_all=1&show_num=22&tag=1&format=json&page=8&callback=newsloadercallback&_=1520314504544'
data=getNewsContent(url)
df=pandas.DataFrame(data)
df.head(2)
df.to_excel('paquwenjianneirong.xls')
```
